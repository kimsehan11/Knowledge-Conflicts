```python
import base64

with open("b64.txt", "r", encoding="utf-8") as f:
    b64 = f.read().strip()

# 혹시 줄바꿈/공백이 섞였으면 제거
b64 = "".join(b64.split())

png_bytes = base64.b64decode(b64)

with open("recovered.png", "wb") as f:
    f.write(png_bytes)

print("saved: recovered.png", len(png_bytes), "bytes")
```
